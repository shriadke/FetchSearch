```python
def clean_text(text):
    import re
    text = str(text).lower()
    text = re.sub('[-]+',' ', text)
    text = re.sub('[^A-Za-z0-9\s]+', '', text)   
    #text = text.strip()
    return text
```

```python
def embd_txt(row):
    ret_string = ""
    #print(row)
    for cols, val in row.items():
        if val == " ":
            continue
        if cols == "RECEIPTS":
            continue
        val = clean_text(val)
        ret_string += str(val+ ", ")
    if ret_string != "":
        ret_string=ret_string[:-2]

    return ret_string
```

```python
brand_offer_category_df["SIMPLE_EMBD"] = brand_offer_category_df.apply(lambda row:embd_txt(row), axis=1)
```

```python
query = "WALMART"
top_k = 5  
q_embeddings =  model.encode([clean_text(query)], show_progress_bar=True)
```

```python
brand_retailer_groups = []
brand_retailer_df =  offer_df[["BRAND", "RETAILER"]]
brand_retailer_df.fillna("", inplace=True)
for brand, group in brand_retailer_df.groupby("BRAND"):
    #print(len(group))
    group = group["RETAILER"].drop_duplicates()
    if len(group) > 1:
        group = list(filter(lambda x: len(x) > 0, group))

    all_retailers = ", ".join(group)
    brand_retailer_groups.append([brand, all_retailers])

brand_retailer_groups = pd.DataFrame(brand_retailer_groups, columns=["BRAND", "ALL_RETAILER"])
brand_retailer_groups
```

```python
retailer_brand_groups = []
brand_retailer_df =  offer_df[["BRAND", "RETAILER"]]
brand_retailer_df.fillna("", inplace=True)
for retailer, group in brand_retailer_df.groupby("RETAILER"):
    #print(len(group))
    if retailer == "":
        continue
    group = group["BRAND"].drop_duplicates()
    if len(group) > 1:
        group = list(filter(lambda x: len(x) > 0, group))

    all_brands = ", ".join(group)
    retailer_brand_groups.append([retailer, all_brands])

retailer_brand_groups = pd.DataFrame(retailer_brand_groups, columns=["RETAILER", "ALL_BRAND"])
retailer_brand_groups
```
